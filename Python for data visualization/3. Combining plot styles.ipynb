```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
```

```python
dataset = randn(100)
sns.distplot(dataset, bins=25)  #Combines KDE with histogram
```

```python
sns.distplot(dataset, bins=25, rug=True, hist=False)
```

```python
#To control specific plots within combined plot
sns.distplot(dataset, bins=25,
             kde_kws={'color':'indianred','label':'KDE PLOT'},
             hist_kws={'color':'blue', 'label':'HIST'})
```

```python
from pandas import Series
ser1 = Series(dataset, name='My_data')
ser1
```

```python
sns.distplot(ser1, bins=25)
```
